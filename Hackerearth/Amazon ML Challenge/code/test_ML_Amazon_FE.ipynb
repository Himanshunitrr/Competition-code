```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import csv
```

```python
#test data
test=pd.read_csv("/content/dataset/test.csv",escapechar = "\\",quoting=csv.QUOTE_NONE)
```

```python
import re
from tqdm import tqdm
import math
```

```python
FE_test=pd.DataFrame()
```

```python
#no. of words
title_words=[]
des_words=[]
bullet_words=[]
for i in tqdm(range(len(test))):

  if(test["TITLE"][i]!=test["TITLE"][i]):
    title_words.append(0)
  else:
    title_words.append(len(test["TITLE"][i].split()))

  if(test["DESCRIPTION"][i]!=test["DESCRIPTION"][i]):
   des_words.append(0)
  else:
   des_words.append(len(test["DESCRIPTION"][i].split()))

  if(test["BULLET_POINTS"][i]!=test["BULLET_POINTS"][i]):
    bullet_words.append(0)
  else:
    bullet_words.append(len(test["BULLET_POINTS"][i].split()))
```

```python
FE_test["title_words"]=title_words
FE_test["des_words"]=des_words
FE_test["bullet_words"]=bullet_words
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
from nltk.tokenize import word_tokenize 
nltk.download('punkt')
```

```python
stop_words = set(stopwords.words('english'))
```

```python
#no. of words without stopwords
title_stopwords=[]
des_stopwords=[]
bullet_stopwords=[]
for i in tqdm(range(len(test))):

  if(test["TITLE"][i]!=test["TITLE"][i]):
    title_stopwords.append(0)
  else:
    word_tokens = word_tokenize(test["TITLE"][i])
    stopwords_x = [w for w in word_tokens if w.lower() in stop_words]
    title_stopwords.append(len(stopwords_x))

  if(test["DESCRIPTION"][i]!=test["DESCRIPTION"][i]):
   des_stopwords.append(0)
  else:
    word_tokens = word_tokenize(test["DESCRIPTION"][i])
    stopwords_x = [w for w in word_tokens if w.lower() in stop_words]
    des_stopwords.append(len(stopwords_x))

  if(test["BULLET_POINTS"][i]!=test["BULLET_POINTS"][i]):
    bullet_stopwords.append(0)
  else:
    word_tokens = word_tokenize(test["BULLET_POINTS"][i])
    stopwords_x = [w for w in word_tokens if w.lower() in stop_words]
    bullet_stopwords.append(len(stopwords_x))
```

```python
FE_test["title_stopwords"]=title_stopwords
FE_test["des_stopwords"]=des_stopwords
FE_test["bullet_stopwords"]=bullet_stopwords
```

```python
#bullet point sentences
bullet_sentences=[]
for i in tqdm(range(len(test))):

  if(test["BULLET_POINTS"][i]!=test["BULLET_POINTS"][i]):
    bullet_sentences.append(0)
  else:
    bullet_sentences.append(len(test["BULLET_POINTS"][i].split(',')))
```

```python
FE_test["bullet_sentences"]=bullet_sentences
```

```python
#average bullet point length
bullet_avg_length=[]
for i in tqdm(range(len(test))):

  if(test["BULLET_POINTS"][i]!=test["BULLET_POINTS"][i]):
    bullet_avg_length.append(0)
  else:
    yy=(test["BULLET_POINTS"][i].split(','))
    sum=0
    for i in range(len(yy)):
      sum+=len(yy[i].split())
    bullet_avg_length.append(sum/len(yy))
```

```python
FE_test["bullet_avg_length"]=bullet_avg_length
```

```python
#total no. of capital characters
title_capital=[]
des_capital=[]
bullet_capital=[]
for i in tqdm(range(len(test))):

  if(test["BULLET_POINTS"][i]!=test["BULLET_POINTS"][i]):
    bullet_capital.append(0)
  else:
    bullet_capital.append(len(re.findall(r'[A-Z]',test["BULLET_POINTS"][i])))

    
  if(test["TITLE"][i]!=test["TITLE"][i]):
    title_capital.append(0)
  else:
    title_capital.append(len(re.findall(r'[A-Z]',test["TITLE"][i])))

  if(test["DESCRIPTION"][i]!=test["DESCRIPTION"][i]):
    des_capital.append(0)
  else:
    des_capital.append(len(re.findall(r'[A-Z]',test["DESCRIPTION"][i])))
```

```python
FE_test["title_capital"]=title_capital
FE_test["des_capital"]=des_capital
FE_test["bullet_capital"]=bullet_capital
```

```python
from textblob import TextBlob
nltk.download('brown')
```

```python
#counting nouns
title_noun=[]
des_noun=[]
bullet_noun=[]
for i in tqdm(range(len(test))):
  if(test["TITLE"][i]!=test["TITLE"][i]):
    title_noun.append(0)
  else:
    blob = TextBlob(test["TITLE"][i])
    title_noun.append(len(blob.noun_phrases))

  if(test["DESCRIPTION"][i]!=test["DESCRIPTION"][i]):
   des_noun.append(0)
  else:
     blob = TextBlob(test["DESCRIPTION"][i])
     des_noun.append(len(blob.noun_phrases))

  if(test["BULLET_POINTS"][i]!=test["BULLET_POINTS"][i]):
   bullet_noun.append(0)
  else:
     blob = TextBlob(test["BULLET_POINTS"][i])
     bullet_noun.append(len(blob.noun_phrases))
```

```python
FE_test["title_noun"]=title_noun
FE_test["des_noun"]=des_noun
FE_test["bullet_noun"]=bullet_noun
```
